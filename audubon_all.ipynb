```python
import pandas as pd
import glob
import os
import shutil 
```

```python
audubon_columns = ['plate_no','common_name','img_file']
audubon_df = pd.DataFrame(columns=audubon_columns)

directory_path = os.getcwd()
files_list = []
file_path = f'{directory_path}/plates'
print(file_path)
for root, directories, files in os.walk(file_path):
    for name in files:
          files_list.append(os.path.join(root, name))
```

```python
audubon_columns = ['plate_no','common_name','file_name','img_file']
audubon_df = pd.DataFrame(columns=audubon_columns)

plate_folders = ['1-99','100-199/','200-299/','300-399/','400-435/']
for plate_folder in plate_folders: 
    audubon_photos_path = directory_path + '/audubon_images/plates/' + plate_folder + '/*'
    all_files = glob.glob(audubon_photos_path)
    for file in all_files:
        file_name = file.rsplit("/",9)[9]
        print(file_name)
        plate_no = "-".join(file_name.split("-", 2)[:2])
        common_name = "-".join(file_name.split("-", 2)[2:])
        common_name = common_name.split(".",1)[0]
        file_info = [plate_no, common_name, file_name, file]
        audubon_df = audubon_df.append(pd.Series(file_info, index=audubon_columns), ignore_index=True)
```

```python
list_of_common_names = audubon_df['common_name'].to_list()
```

```python
audubon_df['in_dataset'] = ''
audubon_df['inaturalist_folder'] = ''
audubon_df['audubon_image'] = ''
for idx, row in audubon_df.iterrows():
    bird_name = row.common_name
    old_path = row.img_file
    new_folder = f'{directory_path}/inaturalist/' + bird_name
    new_path = f'{directory_path}/inaturalist/' + bird_name + '/' + row.file_name
    isFile = os.path.exists(new_folder)
    if isFile == True:
        shutil.copy(old_path, new_path)
        audubon_df.at[idx, 'in_dataset'] = 'Yes'
        audubon_df.at[idx, 'inaturalist_folder'] = new_folder
        audubon_df.at[idx, 'audubon_image'] = new_path
    else:
        #print(bird_name)
        audubon_df.at[idx, 'in_dataset'] = 'No'
pa_birds = audubon_df[audubon_df['in_dataset']=='Yes']
pa_birds = pa_birds.reset_index(drop=True)
pa_birds
```

```python
# Build photomosaic for each bird in Philly using https://github.com/Datadolittle/Photo_Mosaic
import subprocess
pa_birds['mosaic'] = ""
file = open("mosaic_commands.txt","w+")
for i in range(0,pa_birds.shape[0]):
    common_name = pa_birds['common_name'].iloc[i]
    target = pa_birds['img_file'].iloc[i]
    images = pa_birds['inaturalist_folder'].iloc[i]
    output = "mosaics/" + common_name + ".jpeg"
    pa_birds.at[i,'mosaic_img'] = f'{directory_path}/' + output
    command = "python Mosaic_Creator.py --target " + target + " --images " + images + " --grid 100 100 --output " + output
    subprocess.run([command],shell=True)
    file.write(command + '\n')
file.close()
```
