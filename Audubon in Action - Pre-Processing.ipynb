```python
directory_path = os.getcwd()

audubon_columns = ['plate_no','bird_name','audubon_img']
audubon_df = pd.DataFrame(columns=audubon_columns)
print(audubon_df)

plate_folders = ['1-99/','100-199/','200-299/','300-399/','400-435/']
for plate_folder in plate_folders: 
    print(plate_folder)
    audubon_photos_path = '/plates/' + plate_folder + '/*'
    full_path = directory_path + '/plates/' + plate_folder + '/*'
    all_files = glob.glob(full_path)
    print(all_files)
    for file in all_files:
        file_name = file.rsplit("/")[-1]
        print(file_name)
        plate_no = "-".join(file_name.split("-", 2)[:2])
        img_file = 'plates/'+ plate_folder + file_name
        bird_name = "-".join(file_name.split("-", 2)[2:])
        bird_name = bird_name.split(".",1)[0]
        file_info = [plate_no, bird_name, img_file]
        audubon_df = audubon_df.append(pd.Series(file_info, index=audubon_columns), ignore_index=True)
audubon_df
```

## Creating iNaturalist Bird Folders

```python
# Using images from https://www.inaturalist.org/projects/birds-of-philadelphia 
# Filter this dataset for images that are available for use (CC licensing)
# download images from link and save to a folder with the common name of the bird

# Note about this dataset
import pandas as pd
import requests
import os
import numpy as np

inaturalist_data = pd.read_csv("data/observations-218728.csv")
#print(inaturalist_data.shape)
list_of_licenses = ['CC-BY-NC', 'CC-BY-NC-SA', 'CC-BY', 'CC0', 'CC-BY-SA', 'CC-BY-NC-ND']
inaturalist_data = inaturalist_data.loc[inaturalist_data['license'].isin(list_of_licenses)] 
#print(inaturalist_data.shape)
list_of_birds = set(inaturalist_data['common_name'].to_list())
directory_path = os.getcwd()

for bird in list_of_birds: 
    bird_name = bird.lower()
    bird_name = '-'.join(bird_name.split())
    #print(bird, bird_name)
    path = f'{directory_path}/inaturalist/{bird_name}'
    if not os.path.exists(path):
        os.mkdir(f'{directory_path}/inaturalist/{bird_name}')
```

## Download iNaturalist Images

```python
for idx, row in inaturalist_data.iterrows():
    image_url = str(row.image_url)
    common_name = (row.common_name).lower()
    common_name = '-'.join(common_name.split())
    print(common_name)
    file_name = row.id
    img_data = requests.get(image_url).content
    with open(f'inaturalist/{common_name}/{file_name}.jpg', 'wb') as handler:
        print('Getting images...')
        handler.write(img_data) 
```

```python
audubon_df['in_dataset'] = ''
audubon_df['inaturalist_folder'] = ''
#audubon_df['audubon_image'] = ''
for idx, row in audubon_df.iterrows():
    print(bird_name)
    bird_name = row.bird_name
    old_path = row.audubon_img
    new_folder = f'{directory_path}/inaturalist/' + bird_name
    #new_path = '/inaturalist/' + bird_name + '/' + row.file_name #f'{directory_path}/inaturalist/' + bird_name + '/' + row.file_name
    isFile = os.path.exists(new_folder)
    print(new_folder)
    if isFile == True:
        #shutil.copy(old_path, new_path)
        audubon_df.at[idx, 'in_dataset'] = 'Yes'
        audubon_df.at[idx, 'inaturalist_folder'] = 'inaturalist/' + bird_name 
        #audubon_df.at[idx, 'audubon_image'] = old_path
    else:
        #print(bird_name)
        audubon_df.at[idx, 'in_dataset'] = 'No'
pa_birds = audubon_df[audubon_df['in_dataset']=='Yes']
pa_birds = pa_birds.reset_index(drop=True)
pa_birds
```

## Build Mosaics

```python
# Build photomosaic for each bird in Philly using https://github.com/Datadolittle/Photo_Mosaic
import subprocess
pa_birds['mosaic_img'] = ""
#file = open("mosaic_commands.txt","w+")
for i in range(0,pa_birds.shape[0]):
    common_name = pa_birds['bird_name'].iloc[i]
    target = pa_birds['audubon_img'].iloc[i]
    target = f'{directory_path}/' + target
    images = pa_birds['inaturalist_folder'].iloc[i]
    images = f'{directory_path}/' + images
    output = "mosaics/" + common_name + ".jpeg"
    pa_birds['mosaic_img'].loc[pa_birds.index[i]] = output
    #pa_birds['mosaic_img'] = output
    output = f'{directory_path}/' + output
    #pa_birds.at[i,'mosaic_img'] = output
    print(common_name, output)
    command = "python Mosaic_Creator.py --target " + target + " --images " + images + " --grid 100 100 --output " + output
    subprocess.run([command],shell=True)
    #file.write(command + '\n')
#file.close()
```

## Create Instagram Posts

```python
# Create an Instagram Post with the data - this will be used outside of pre-processing
from PIL import Image  
from PIL import ImageFont
from PIL import ImageDraw
import datetime
import re

width = 1080
height = 1080

pa_birds['display_img'] = ''
for idx, row in pa_birds.iterrows():
    first_line = row.common_name
    second_line = row.scientific_name
    third_line = "Observed " + str(row['count']) + " times since June 19, 2020"
    
    obs = row.observed_on # double check that this works
    obs = datetime.datetime.strptime(obs,'%Y-%m-%d')
    obs = obs.strftime("%B %e, %Y")
    obs = re.sub(' +', ' ', obs)
    #print(obs)
    
    fourth_line = "Last spotted " + str(obs) + "/n" + "near " + row.place_guess
    fifth_line = "Created by Emily Esten for Audubon in Action at the Earth Week Data Jam 2022"
    print(row.audubon_img)
    print(row.mosaic_img)
    orig_image = f'{directory_path}/' + row.audubon_img
    created_image = f'{directory_path}/' + row.mosaic_img
    output = f'{directory_path}/displays/' + row.bird_name + ".jpeg"
    pa_birds.at[idx, 'display_img'] = 'displays/' + row.bird_name + ".jpeg"

    img  = Image.new(mode = "RGB", size = (width, height), color = (255,255,255))
    img1 = Image.open(orig_image)
    img2 = Image.open(created_image)
    img1 = img1.resize((400,400))
    img2 = img2.resize((400,400))
    img.paste(img1, (0,0))
    img.paste(img2, (680,0))
    
    draw = ImageDraw.Draw(img)
    font = ImageFont.truetype(f"{directory_path}/fonts/Roboto/Roboto-Regular.ttf", size=12)
    credit = "Courtesy of the John James Audubon Center at Mill Grove,\nMontgomery County Audubon Collection, and Zebra Publishing"
    draw.text((180,450), credit, (0,0,0), font=font, anchor="mm", align="left")
    
    font = ImageFont.truetype(f"{directory_path}/fonts/Roboto/Roboto-Bold.ttf", size=52)
    draw.text((540,520), first_line, (0,0,0), font=font, anchor="mm", align="center")

    font = ImageFont.truetype(f"{directory_path}/fonts/Roboto/Roboto-Italic.ttf", size=32)
    draw.text((540,580), second_line, (0,0,0), font=font, anchor="mm", align="center")

    font = ImageFont.truetype(f"{directory_path}/fonts/Roboto/Roboto-Regular.ttf", size=22)
    draw.text((540,680), third_line, (0,0,0), font=font, anchor="mm", align="center")

    font = ImageFont.truetype(f"{directory_path}/fonts/Roboto/Roboto-Regular.ttf", size=22)
    draw.text((540,720), fourth_line, (0,0,0), font=font, anchor="mm", align="center")

    font = ImageFont.truetype(f"{directory_path}/fonts/Roboto/Roboto-Regular.ttf", size=12)
    draw.text((540,1000), fifth_line, (0,0,0), font=font, anchor="mm", align="center")

    img.save(output)
```

```python
## Create Sound Files
import pandas as pd
import numpy as np
import os

inaturalist_data = pd.read_csv("data/observations-218728.csv")
print(inaturalist_data.shape)
list_of_licenses = ['CC-BY-NC', 'CC-BY-NC-SA', 'CC-BY', 'CC0', 'CC-BY-SA', 'CC-BY-NC-ND']
inaturalist_data = inaturalist_data.loc[inaturalist_data['license'].isin(list_of_licenses)] 
bird_count = inaturalist_data.groupby(['common_name']).size().reset_index()
#bird_count.to_csv("bird_count.csv")
sounds = inaturalist_data.dropna(subset=['sound_url'])
sounds.reset_index(drop=True)
print(sounds.shape)
```

```python
birds_cant_find = []
#pa_birdy = pa_birds[pa_birds['bird_name'].isin(empty)]
count = 0
for idx, row in pa_birds.iterrows():
        sounds_columns = ['common_name','file','license','recordist','XC-number','stable_url','file-name']
        sounds_df = pd.DataFrame(columns=sounds_columns)
        sounds_folder = row.sounds_folder
        common_name = row.common_name
        bird_name = row.bird_name
        print(common_name)
        if bird_name not in empty:
            print("done!")
        else:
            common_name = common_name.lower()
            query_url = urllib.parse.quote_plus(common_name)
            xeno = "https://xeno-canto.org/api/2/recordings?query=" + query_url
            json_data = requests.get(xeno).json()
            for file in json_data['recordings']:
                file_info = [common_name, file['file'], file['lic'], file['rec'], file['id'], file['url'], file['file-name']]
                sounds_df = sounds_df.append(pd.Series(file_info, index=sounds_columns), ignore_index=True)
            licenses = ['//creativecommons.org/licenses/by-nc-sa/4.0/', '//creativecommons.org/licenses/by-nc-sa/3.0/','//creativecommons.org/licenses/by-sa/3.0/']
            sounds_df = sounds_df[sounds_df['license'].isin(licenses)].reset_index()
            sounds_df.to_csv(f'sounds/{bird_name}.csv', index=False)
            if sounds_df.shape[0] != 0:
                sound_url = sounds_df['file'].loc[sounds_df.index[0]]
                sound_data = requests.get(sound_url).content
                sound_type = sounds_df['file-name'].loc[sounds_df.index[0]].split('.')[-1]
                print(sound_type)
                file_name = f'{bird_name}_' + str(count)
                count+=1
                print(file_name)
                with open(f'{directory_path}/sounds/{bird_name}/{file_name}.{sound_type}', 'wb') as handler:
                    print('Getting sounds...')
                    handler.write(sound_data)
            else:
                print("no sounds found for " + bird_name)
                no_bird = common_name.title()
                birds_cant_find.append(no_bird)
                
        count +=1
```

```python
from pydub import AudioSegment
import os, random
import pandas as pd
import numpy as np
import os
import glob

import datetime
from datetime import date
directory_path = os.getcwd()

inaturalist_data = pd.read_csv("inaturalist/observations-218728.csv")
list_of_licenses = ['CC-BY-NC', 'CC-BY-NC-SA', 'CC-BY', 'CC0', 'CC-BY-SA', 'CC-BY-NC-ND']
inaturalist_data = inaturalist_data.loc[inaturalist_data['license'].isin(list_of_licenses)].reset_index() 
birds_to_include = pa_birds['common_name'].to_list()
inaturalist_data = inaturalist_data[inaturalist_data['common_name'].isin(birds_to_include)]

birds_remove = []
for bird in birds_cant_find:
    result = bird.title()
    print(result)
    birds_remove.append(bird)
inaturalist_data = inaturalist_data[~inaturalist_data['common_name'].isin(birds_remove)]

inaturalist_data.to_csv("data/birds_for_sound.csv")
dates = inaturalist_data['observed_on'].to_list()
start_date = datetime.datetime.strptime("2010-6-19",'%Y-%m-%d')
days_dates = {}
for day in dates: 
    end_date = datetime.datetime.strptime(day, '%Y-%m-%d')
    delta = end_date - start_date
    days_dates[delta.days] = day
```

```python
from pydub.playback import play
res = int(list(days_dates.keys())[0])
all_sounds = AudioSegment.silent(duration=1)
all_sounds.duration_seconds == int(delta.days)

for i in range(res,int(delta.days)):
    print(i)
    if i in days_dates.keys():
        print("yes")
        value = days_dates[i]
        spotted_birds = inaturalist_data[inaturalist_data['observed_on']==value].reset_index()
        print(spotted_birds.shape[0])
        common_name = spotted_birds['common_name'].loc[spotted_birds.index[0]]
        print(common_name)
        bird_name = common_name.lower()
        bird_name = '-'.join(bird_name.split())
        sounds_folder = f'{directory_path}/sounds/' + bird_name
        sound_file = random.choice(os.listdir(sounds_folder))
        sound_type = sound_file.rsplit('.')[1]
        sound_file = sounds_folder + "/" + sound_file
        print(sound_file)
        sound = AudioSegment.from_file(sound_file, sound_type)
        sound = sound[1000:2000]
        combined = sound + 10
        for idx, row in spotted_birds[1:].iterrows(): 
            common_name = spotted_birds['common_name'].loc[spotted_birds.index[idx]]
            bird_name = common_name.lower()
            bird_name = '-'.join(bird_name.split())
            sounds_folder = f'{directory_path}/sounds/' + bird_name
            sound_file = random.choice(os.listdir(sounds_folder))
            sound_type = sound_file.rsplit('.')[1]
            sound_file = sounds_folder + "/" + sound_file
            sound = AudioSegment.from_file(sound_file, sound_type)
            sound = sound[1000:2000]
            sound = sound + 15
            combined.overlay(sound)
            #play(combined)
    else:
        print("no")
        combined = AudioSegment.silent(duration=1000)
    all_sounds = all_sounds + combined
all_sounds.export("sounds/output.mp3", format="mp3")
print("sound created!")
```
